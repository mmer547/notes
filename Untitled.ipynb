```python
fig = plt.figure(figsize=(5,6))

ims = []

ro = 50
ri = 40
ror = ro - ri
L = 80

width = 5

angle = []

for i in range(360):
    theta = [i for i in range(360)]
    
    x = [ro * np.cos(i/180*math.pi) for i in theta]
    y = [ro * np.sin(i/180*math.pi) for i in theta]
    im = plt.plot(x, y)
    
    x2 = [ri * np.cos(i/180*math.pi) for i in theta]
    y2 = [ri * np.sin(i/180*math.pi) for i in theta]
    x2 = x2 + ror * np.cos(-i/180*math.pi)
    y2 = y2 + ror * np.sin(-i/180*math.pi)
    im2 = plt.plot(x2, y2)
    
    x3 = ror * np.cos(-i/180*math.pi)
    y3 = ror * np.sin(-i/180*math.pi)
    x4 = -ror * np.cos(i/180*math.pi)
    y4 = ror * np.sin(-i/180*math.pi) + L
    im3 = plt.plot([x3, x4], [y3, y4])
    im4 = plt.plot([x3, x4], [y3, y4], "o")

    x5 = [ror * np.cos(i/180*math.pi) for i in theta]
    y5 = [ror * np.sin(i/180*math.pi) for i in theta]
    im5 = plt.plot(x5, y5)
    
    x6 = [ror * np.cos(i/180*math.pi) for i in theta]
    y6 = [ror * np.sin(i/180*math.pi)+L for i in theta]
    im6 = plt.plot(x6, y6)
    
    x7 = ror * np.cos(-i/180*math.pi) + width/2
    y7 = ror * np.sin(-i/180*math.pi)
    x8 = -ror * np.cos(i/180*math.pi) + width/2
    y8 = ror * np.sin(-i/180*math.pi) + L 
    im7 = plt.plot([x7, x8], [y7, y8])
    
    x9 = ror * np.cos(-i/180*math.pi) + -width/2
    y9 = ror * np.sin(-i/180*math.pi)
    x10 = -ror * np.cos(i/180*math.pi) + -width/2
    y10 = ror * np.sin(-i/180*math.pi) + L 
    im8 = plt.plot([x9, x10], [y9, y10])
    
#     angle.append(math.acos((y4-y3)/L))
    angle.append((y4-y3))
    
    ims.append(im+im2+im3+im4+im5+im6+im7+im8)

ani = animation.ArtistAnimation(fig, ims, interval=10)
plt.show()
```
